```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("Tabla_sexo_clase_sobrevivientes.csv", index_col="Clases")

muestra = df[:10].copy()

colors = ["#6495ED", "#DE3163"]

muestra.plot(kind = 'bar', stacked="True" ,width=0.8, figsize=(8,8),color = colors,);

plt.title("Cantidad de Sobrevivientes por Clase")
plt.ylabel("Cantidad de Pasajeros")
```
